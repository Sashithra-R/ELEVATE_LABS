DAY - 1 DATA CLEANING AND PREPROCESSING

Using pandas to clean and preprocess the dataset

```python
#importing the necessary library
import pandas as pd
```

```python
#loading the dataset
df = pd.read_csv("/content/netflix_titles.csv")
df
```

```python
#renaming the column headers to be clean and uniform
df.columns = df.columns.str.lower().str.strip().str.replace(' ','_')
```

```python
#checking for basic info like missing values
df.info()
```

```python
#handling the missing values
df['director'] = df['director'].fillna('Not Available')
df['cast'] = df['cast'].fillna('Not Available')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unrated')
df['duration'] = df['duration'].fillna('Unknown')
```

```python
#converting and cleaning date_added column
df['date_added']=pd.to_datetime(df['date_added'],errors = 'coerce')

#extract the month and year
df['year_added']=df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

#dropping the date_added column
df = df.drop(columns=['date_added'])
```

```python
#Cleaning and splitting duration column
df[['duration_int','duration_type']] = df['duration'].str.extract('(\d+)\s*(\w+)')
df['duration_int']=pd.to_numeric(df['duration_int'],errors = 'coerce')

#dropping the duration column
df = df.drop(columns = ['duration'])
```

```python
#replacing the null values in the splitted columns
df['year_added'] = df['year_added'].fillna(0).astype(int)
df['month_added'] = df['month_added'].fillna(0).astype(int)
df['duration_int'] = df['duration_int'].fillna(0).astype(int)
df['duration_type'] = df['duration_type'].fillna('Unknown')
```

```python
#removing duplicates and resetting index
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

```python
#once again checking for the null values after cleaning
df.info()
```

```python
#saving the cleaned dataset
df.to_csv('cleaned_netflix_data.csv',index=False)
df.head()
```
